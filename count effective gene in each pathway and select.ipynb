```python
import pandas as pd
import numpy as np
```

```python
counts = pd.read_csv('all_genes.csv')
```

```python
counts.head()
```

```python
mapping = dict(counts[['gene_sliced', 'Approved_Symbol']].values)
```

```python
gene_symbol = list(counts['Approved_Symbol'])
gene_symbol
```

```python
pathways = pd.read_csv('pathways_gene_approved symbol form.csv')
```

```python
pathways.head()
```

```python
from itertools import repeat
count = list(repeat(0,len(pathways['pathways'])))
```

```python
pathway_effectivegene = pd.DataFrame()
```

```python
se = pd.Series(pathways['pathways'])
pathway_effectivegene['pathways']=se.values
```

```python
list_name = []
for i in range(400):
    list_name.append('gene'+ str(i + 1))

for gene in list_name:
    pathway_effectivegene[gene] = 0
```

```python
pathway_effectivegene.head()
```

```python
pathway_effectivegene.iloc[1,1]=pathways.iloc[1,1]
pathway_effectivegene.head()
```

```python
for i in range(len(pathways['pathways'])):
    for j in range(400):
        if pathways.iloc[i,j] in gene_symbol:
            count[i] += 1
            pathway_effectivegene.iloc[i,count[i]]=pathways.iloc[i,j]
            
        
    
```

```python
pathways['effective gene counts(appears in cpm file)'] = count
```
