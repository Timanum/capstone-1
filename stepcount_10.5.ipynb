```python
import os

directory_list = []
for root, dirs, files in os.walk("/Users/yiran/Desktop/Step-Omics/step count data files", topdown=False):
    for name in dirs:
        directory_list.append(os.path.join(root, name))

import glob
import pandas as pd

step_count_activity=pd.DataFrame()
filenames=pd.DataFrame()
str='/*.xls'
paths=[x+str for x in directory_list]
paths.sort()
```

```python
for path in paths:
    filenames = glob.glob(path)
    for filename in filenames:
        df=pd.read_excel(filename, sheet_name='Activities')
        df['studyID']=path[54:58]
        step_count_activity=step_count_activity.append(df)
step_count_activity=step_count_activity.set_index('studyID')
step_count_activity
```
